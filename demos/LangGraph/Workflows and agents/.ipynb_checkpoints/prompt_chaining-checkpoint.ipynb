```python
from typing import TypedDict
```

```python
from langgraph.graph import StateGraph, START, END
```

```python
from initialize_the_llm import llm
```

```python
class State(TypedDict):
    topic: str
    joke: str
    improved_joke: str
    final_joke: str
```

```python
def generate_joke(state: State) -> dict:
    """È¶ñÊ¨°Ë∞ÉÁî®LLMÁîüÊàêÊúÄÂàùÁöÑÁ¨ëËØù"""
    
    msg = llm.invoke(f"ÂÜô‰∏Ä‰∏™ÂÖ≥‰∫é{state['topic']}ÁöÑÁ¨ëËØù„ÄÇ")
    return {"joke": msg.content}
```

```python
def check_punchline(state: State) -> str:
    """Ê£ÄÊü•Á¨ëËØùÊòØÂê¶ÊúâÁ¨ëÁÇπÁöÑÈó®ÊéßÂáΩÊï∞"""

    if "Ôºü" in state['joke'] or "ÔºÅ" in state['joke']:
        return "Pass"
    return "Fail"
```

```python
def improve_joke(state: State) -> dict:
    """ÂÜçÊ¨°Ë∞ÉÁî®LLMÊîπËøõÁ¨ëËØù"""

    msg = llm.invoke(f"ÈÄöËøáÊ∑ªÂä†ÊñáÂ≠óÊ∏∏Êàè‰ΩøËøô‰∏™Á¨ëËØùÊõ¥ÊúâË∂£Ôºö{state['joke']}")
    return {"improved_joke": msg.content}
```

```python
def polish_joke(state: State) -> dict:
    """Á¨¨‰∏âÊ¨°Ë∞ÉÁî®LLMÂØπÊîπËøõÁöÑÁ¨ëËØùËøõ‰∏ÄÊ≠•Ê∂¶Ëâ≤"""

    msg = llm.invoke(f"ÁªôËøô‰∏™Á¨ëËØùÂä†‰∏Ä‰∏™Âá∫‰∫∫ÊÑèÂ§ñÁöÑËΩ¨ÊäòÔºö{state['improved_joke']}")
    return {"final_joke": msg.content}
```

```python
workflow = StateGraph(State)
```

```python
workflow.add_node("generate_joke", generate_joke)
```

```python
workflow.add_node("improve_joke", improve_joke)
```

```python
workflow.add_node("polish_joke", polish_joke)
```

```python
workflow.add_edge(START, "generate_joke")
```

```python
workflow.add_conditional_edges("generate_joke", check_punchline, {"Pass": END, "Fail": "improve_joke"})
```

```python
workflow.add_edge("improve_joke", "polish_joke") 
```

```python
workflow.add_edge("polish_joke", END)
```

```python
chain = workflow.compile()
```
