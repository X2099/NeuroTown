```python
class SearchQuery(BaseModel):
    search_query: str = Field(None, description="针对网络搜索进行优化的查询。")
    justification: str = Field(None, description="此查询与用户的请求相关的理由。")
```

```python
structured_llm = llm.with_structured_output(SearchQuery)
```

```python
output = structured_llm.invoke("钙CT值与高胆固醇有何关系？")
```
